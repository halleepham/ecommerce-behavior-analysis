```python
import pandas as pd
df = pd.read_parquet("/kaggle/input/datasets/kylenaluan/user-segments")
sf_df = pd.read_parquet("/kaggle/input/datasets/kylenaluan/session-features")
```

```python
df.head()
```

```python
sf_df.head()
```

```python
merged = sf_df.merge(df, on='user_id', how='left')
```

```python
merged
```

```python
new_merged = (
    merged
    .groupby(['user_id', 'cluster_label'])
    .agg({
        'n_events': 'sum',
        'unique_products': 'sum',
        'n_views': 'sum',
        'n_carts': 'sum',
        'n_purchases': 'sum',
        'session_revenue': 'sum',
        'duration_seconds': 'sum',
    })
    .reset_index()
)

new_merged
```

Analysis can describe differences between groups with statistical rigor, but cannot establish that any one factor caused those differences. All conclusions are associational, not causal. This limitation is explicitly acknowledged in the notebook.

Questions
Do passive browsers view more items?

Do active browsers have a higher cart rate/purchase rate than passive browsers?

Do buyers (one-time and engaged) have a higher purchase rate than the browser clusters?

Do buyers spend as much time on the website?

Do active browsers spend more money than buyers?

Question 1:
# Is there a difference in view counts between the clusters?
Null Hypothesis: All four groups have the same view counts distribution.

Alternative Hypothesis: At least one of the four has a differing view count distribution.

```python
from scipy.stats import ttest_ind

# Split groups
active = new_merged[new_merged['cluster_label'] == 'Active Browsers']['n_views']
passive = new_merged[new_merged['cluster_label'] == 'Passive Browsers']['n_views']
one_time_buyers = new_merged[new_merged['cluster_label'] == 'One Time Buyers']['n_views']
engaged_buyers = new_merged[new_merged['cluster_label'] == 'Engaged Buyers']['n_views']

from scipy.stats import mannwhitneyu

stat1, p_value1 = mannwhitneyu(active, passive, alternative='two-sided')
stat2, p_value2 = mannwhitneyu(active, one_time_buyers, alternative='two-sided')
stat3, p_value3 = mannwhitneyu(active, engaged_buyers, alternative='two-sided')
stat4, p_value4 = mannwhitneyu(one_time_buyers, passive, alternative='two-sided')
stat5, p_value5 = mannwhitneyu(engaged_buyers, passive, alternative='two-sided')
stat6, p_value6 = mannwhitneyu(one_time_buyers, engaged_buyers, alternative='two-sided')

print("Active-Passive T-Statistic:", stat1)
print("Active-Passive P-value:", p_value1)
print("Active-OneTimeBuyer T-Statistic:", stat2)
print("Active-OneTimeBuyer P-value:", p_value2)
print("Active-EngagedBuyer T-Statistic:", stat3)
print("Active-EngagedBuyer P-value:", p_value3)
print("OneTimeBuyer-Passive T-Statistic:", stat4)
print("OneTimeBuyer-Passive P-value:", p_value4)
print("EngagedBuyer-Passive T-Statistic:", stat5)
print("EngagedBuyer-Passive P-value:", p_value5)
print("OneTimeBuyer-EngagedBuyer T-Statistic:", stat6)
print("OneTimeBuyer-EngagedBuyer P-value:", p_value6)
```

```python
print("Active Browser's median views:", active.median())
print("Passive Browser's median views:", passive.median())
print("One Time Buyer's median views:", one_time_buyers.median())
print("Engaged Buyer's median views:", engaged_buyers.median())
```

Conducted a non-parametric Mann-Whitney U test because the distribution was not assumed to be normal. The p-values for each difference was less than 0.05 so each cluster's view count distribution is significantly different. 

# Is there a difference in cart rate between clusters?
Null Hypothesis: All four groups have the same cart rates.

Alternative Hypothesis: At least one of the four has a differing cart rate.

```python
from statsmodels.stats.proportion import proportions_ztest

grouped = new_merged.groupby('cluster_label').agg({
    'n_carts': 'sum',
    'n_views': 'sum'
})

active_cart = grouped.loc['Active Browsers', 'n_carts']
active_total   = grouped.loc['Active Browsers', 'n_views']

passive_cart = grouped.loc['Passive Browsers', 'n_carts']
passive_total   = grouped.loc['Passive Browsers', 'n_views']

one_time_buyer_cart = grouped.loc['One Time Buyers', 'n_carts']
one_time_buyer_total   = grouped.loc['One Time Buyers', 'n_views']

engaged_buyer_cart = grouped.loc['Engaged Buyers', 'n_carts']
engaged_buyer_total   = grouped.loc['Engaged Buyers', 'n_views']

stat1, p_value1 = proportions_ztest(
    [active_cart, passive_cart],
    [active_total, passive_total]
)

stat2, p_value2 = proportions_ztest(
    [active_cart, one_time_buyer_cart],
    [active_total, one_time_buyer_total]
)

stat3, p_value3 = proportions_ztest(
    [active_cart, engaged_buyer_cart],
    [active_total, engaged_buyer_total]
)

stat4, p_value4 = proportions_ztest(
    [one_time_buyer_cart, passive_cart],
    [one_time_buyer_total, passive_total]
)

stat5, p_value5 = proportions_ztest(
    [engaged_buyer_cart, passive_cart],
    [engaged_buyer_total, passive_total]
)

stat6, p_value6 = proportions_ztest(
    [one_time_buyer_cart, engaged_buyer_cart],
    [one_time_buyer_total, engaged_buyer_total]
)

print(f"Active-Passive Z-stat: {stat1:.3f}")
print(f"Active-Passive P-value: {p_value1:.4f}")

print(f"Active-OneTimeBuyer Z-stat: {stat2:.3f}")
print(f"Active-OneTimeBuyer P-value: {p_value2:.4f}")

print(f"Active-EngagedBuyer Z-stat: {stat3:.3f}")
print(f"Active-EngagedBuyer P-value: {p_value3:.4f}")

print(f"OneTimeBuyer-Passive Z-stat: {stat4:.3f}")
print(f"OneTimeBuyer-Passive P-value: {p_value4:.4f}")

print(f"EngagedBuyer-Passive Z-stat: {stat5:.3f}")
print(f"EngagedBuyer-Passive P-value: {p_value5:.4f}")

print(f"OneTimeBuyer-EngagedBuyer Z-stat: {stat6:.3f}")
print(f"OneTimeBuyer-EngagedBuyer P-value: {p_value6:.4f}")
```

```python
print(f"Active Proportion: {active_cart}/{active_total} = {active_cart/active_total}")
print(f"Passive Proportion: {passive_cart}/{passive_total} = {passive_cart/passive_total}")
print(f"One Time Buyers Proportion: {one_time_buyer_cart}/{one_time_buyer_total} = {one_time_buyer_cart/one_time_buyer_total}")
print(f"Engaged Buyers Proportion: {engaged_buyer_cart}/{engaged_buyer_total} = {engaged_buyer_cart/engaged_buyer_total}")
```

Conducted a proportion z-test because we want to compare probability outcome rather than distributions. The p-values for each difference was less than 0.05 so each cluster's probability of carting an item is significantly different. 

# Is there a difference in purchase rate between the clusters?

Null Hypothesis: All four groups have the same purchase rates.

Alternative Hypothesis: At least one of the four has a differing purchase rate.

```python
from statsmodels.stats.proportion import proportions_ztest

grouped = new_merged.groupby('cluster_label').agg({
    'n_purchases': 'sum',
    'n_views': 'sum'
})

active_purchase = grouped.loc['Active Browsers', 'n_purchases']
active_total   = grouped.loc['Active Browsers', 'n_views']

passive_purchase = grouped.loc['Passive Browsers', 'n_purchases']
passive_total   = grouped.loc['Passive Browsers', 'n_views']

one_time_buyer_purchase = grouped.loc['One Time Buyers', 'n_purchases']
one_time_buyer_total   = grouped.loc['One Time Buyers', 'n_views']

engaged_buyer_purchase = grouped.loc['Engaged Buyers', 'n_purchases']
engaged_buyer_total   = grouped.loc['Engaged Buyers', 'n_views']

stat1, p_value1 = proportions_ztest(
    [active_purchase, passive_purchase],
    [active_total, passive_total]
)

stat2, p_value2 = proportions_ztest(
    [active_purchase, one_time_buyer_purchase],
    [active_total, one_time_buyer_total]
)

stat3, p_value3 = proportions_ztest(
    [active_purchase, engaged_buyer_purchase],
    [active_total, engaged_buyer_total]
)

stat4, p_value4 = proportions_ztest(
    [one_time_buyer_purchase, passive_purchase],
    [one_time_buyer_total, passive_total]
)

stat5, p_value5 = proportions_ztest(
    [engaged_buyer_purchase, passive_purchase],
    [engaged_buyer_total, passive_total]
)

stat6, p_value6 = proportions_ztest(
    [one_time_buyer_purchase, engaged_buyer_purchase],
    [one_time_buyer_total, engaged_buyer_total]
)

print(f"Active-Passive Z-stat: {stat1:.3f}")
print(f"Active-Passive P-value: {p_value1:.4f}")

print(f"Active-OneTimeBuyer Z-stat: {stat2:.3f}")
print(f"Active-OneTimeBuyer P-value: {p_value2:.4f}")

print(f"Active-EngagedBuyer Z-stat: {stat3:.3f}")
print(f"Active-EngagedBuyer P-value: {p_value3:.4f}")

print(f"OneTimeBuyer-Passive Z-stat: {stat4:.3f}")
print(f"OneTimeBuyer-Passive P-value: {p_value4:.4f}")

print(f"EngagedBuyer-Passive Z-stat: {stat5:.3f}")
print(f"EngagedBuyer-Passive P-value: {p_value5:.4f}")

print(f"OneTimeBuyer-EngagedBuyer Z-stat: {stat6:.3f}")
print(f"OneTimeBuyer-EngagedBuyer P-value: {p_value6:.4f}")
```

```python
print(f"Active Proportion: {active_purchase}/{active_total} = {active_purchase/active_total}")
print(f"Passive Proportion: {passive_purchase}/{passive_total} = {passive_purchase/passive_total}")
print(f"One Time Buyers Proportion: {one_time_buyer_purchase}/{one_time_buyer_total} = {one_time_buyer_purchase/one_time_buyer_total}")
print(f"Engaged Buyers Proportion: {engaged_buyer_purchase}/{engaged_buyer_total} = {engaged_buyer_purchase/engaged_buyer_total}")
```

Conducted a proportion z-test because we want to compare probability outcome rather than distributions. The p-values for each difference was less than 0.05 so each cluster's probability of carting an item is significantly different.
